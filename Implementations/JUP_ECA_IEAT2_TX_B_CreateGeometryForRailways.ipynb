```python
import sys, os, importlib, logging

import pandas as pd
import geopandas as gpd

from shapely.geometry import Point, LineString
```

```python
in_nodes_file = r"S:\REGIONAL\ECA\Projects\Rail_Network_Mapping\data\rail_nodes.csv"
in_edges_file = r"S:\REGIONAL\ECA\Projects\Rail_Network_Mapping\data\edges.csv"
```

```python
# Read in nodes and edges
inD = pd.read_csv(in_nodes_file)
inE = pd.read_csv(in_edges_file)
# Create point objects out of nodes
inD['geometry'] = [Point(x, y) for x, y in zip(inD['X'], inD['Y'])]
```

```python
# Merge node points onto edges, once for source and once for target
inE_source = inE.merge(inD.loc[:,["node_id",'name','geometry']], 
          left_on="source", right_on="node_id")
inE_source = inE_source.merge(inD.loc[:,["node_id",'name','geometry']], 
          left_on="target", right_on="node_id")
```

```python
inE_geom = [LineString([x,y]) for x,y in zip (inE_source['geometry_x'], inE_source['geometry_y'])]
```

```python
inE_final = inE_source.loc[:,["source","target","link_id","length","time","name_x"]]
inE_final = gpd.GeoDataFrame(inE_final, geometry=inE_geom, crs={'init':'epsg:4326'})
inE_final.to_file(in_edges_file.replace(".csv", ".shp"))
```
